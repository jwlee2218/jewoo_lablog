```python
# Fill NaN with According Groups
# Masters
tr_master_passenger = train_data[train_data["Name"].str.contains("Master")]
print("Master_tr:",tr_master_passenger["Age"].mean())
tr_master_passenger["Age"].fillna(tr_master_passenger["Age"].mean(),inplace = True)
train_data.update(tr_master_passenger)

t_master_passenger = test_data[test_data["Name"].str.contains("Master")]
print("Master_t:",t_master_passenger["Age"].mean())
t_master_passenger["Age"].fillna(t_master_passenger["Age"].mean(),inplace = True)
test_data.update(t_master_passenger)

# Doctors
tr_dr_passenger = train_data[train_data["Name"].str.contains("Dr")]
print("Dr:",tr_dr_passenger["Age"].mean())
tr_dr_passenger["Age"].fillna(tr_dr_passenger["Age"].mean(),inplace = True)
train_data.update(tr_dr_passenger)

t_dr_passenger = test_data[test_data["Name"].str.contains("Dr")]
print("Dr:",t_dr_passenger["Age"].mean())
t_dr_passenger["Age"].fillna(t_dr_passenger["Age"].mean(),inplace = True)
test_data.update(t_dr_passenger)

# Male
mr_passenger = train_data[train_data["Name"].str.contains("Mr")]

mr_1_C = mr_passenger[mr_passenger["Pclass"]==1][mr_passenger["Embarked"]=="C"]
mr_1_S = mr_passenger[mr_passenger["Pclass"]==1][mr_passenger["Embarked"]=="S"]
mr_2_C = mr_passenger[mr_passenger["Pclass"]==2][mr_passenger["Embarked"]=="C"]
mr_2_S = mr_passenger[mr_passenger["Pclass"]==2][mr_passenger["Embarked"]=="S"]
mr_3_C = mr_passenger[mr_passenger["Pclass"]==3][mr_passenger["Embarked"]=="C"]
mr_3_S = mr_passenger[mr_passenger["Pclass"]==3][mr_passenger["Embarked"]=="S"]
mr_3_Q = mr_passenger[mr_passenger["Pclass"]==3][mr_passenger["Embarked"]=="Q"]

mr_1_C["Age"].fillna(mr_1_C["Age"].mean(), inplace = True)
mr_1_S["Age"].fillna(mr_1_S["Age"].mean(), inplace = True)
mr_2_C["Age"].fillna(mr_2_C["Age"].mean(), inplace = True)
mr_2_S["Age"].fillna(mr_2_S["Age"].mean(), inplace = True)
mr_3_C["Age"].fillna(mr_3_C["Age"].mean(), inplace = True)
mr_3_S["Age"].fillna(mr_3_S["Age"].mean(), inplace = True)
mr_3_Q["Age"].fillna(mr_3_Q["Age"].mean(), inplace = True)

t_mr_passenger = test_data[test_data["Name"].str.contains("Mr")]

t_mr_1_C = t_mr_passenger[t_mr_passenger["Pclass"]==1][t_mr_passenger["Embarked"]=="C"]
t_mr_1_S = t_mr_passenger[t_mr_passenger["Pclass"]==1][t_mr_passenger["Embarked"]=="S"]
t_mr_2_C = t_mr_passenger[t_mr_passenger["Pclass"]==2][t_mr_passenger["Embarked"]=="C"]
t_mr_2_S = t_mr_passenger[t_mr_passenger["Pclass"]==2][t_mr_passenger["Embarked"]=="S"]
t_mr_2_Q = t_mr_passenger[t_mr_passenger["Pclass"]==2][t_mr_passenger["Embarked"]=="Q"]
t_mr_3_C = t_mr_passenger[t_mr_passenger["Pclass"]==3][t_mr_passenger["Embarked"]=="C"]
t_mr_3_S = t_mr_passenger[t_mr_passenger["Pclass"]==3][t_mr_passenger["Embarked"]=="S"]
t_mr_3_Q = t_mr_passenger[t_mr_passenger["Pclass"]==3][t_mr_passenger["Embarked"]=="Q"]


t_mr_1_C["Age"].fillna(t_mr_1_C["Age"].mean(), inplace = True)
t_mr_1_S["Age"].fillna(t_mr_1_S["Age"].mean(), inplace = True)
t_mr_2_C["Age"].fillna(t_mr_2_C["Age"].mean(), inplace = True)
t_mr_2_S["Age"].fillna(t_mr_2_S["Age"].mean(), inplace = True)
t_mr_2_Q["Age"].fillna(t_mr_2_Q["Age"].mean(), inplace = True)
t_mr_3_C["Age"].fillna(t_mr_3_C["Age"].mean(), inplace = True)
t_mr_3_S["Age"].fillna(t_mr_3_S["Age"].mean(), inplace = True)
t_mr_3_Q["Age"].fillna(t_mr_3_Q["Age"].mean(), inplace = True)


train_data.update(mr_1_C)
train_data.update(mr_2_C)
train_data.update(mr_1_S)
train_data.update(mr_2_S)
train_data.update(mr_3_C)
train_data.update(mr_3_S)
train_data.update(mr_3_Q)

test_data.update(t_mr_1_C)
test_data.update(t_mr_2_C)
test_data.update(t_mr_1_S)
test_data.update(t_mr_2_S)
test_data.update(t_mr_2_Q)
test_data.update(t_mr_3_C)
test_data.update(t_mr_3_S)
test_data.update(t_mr_3_Q)


# Female
## Mrs.
mrs_passenger = train_data[train_data["Name"].str.contains("Mrs")]

mrs_1_C = mrs_passenger[mrs_passenger["Pclass"]==1][mrs_passenger["Embarked"]=="C"]
mrs_1_S = mrs_passenger[mrs_passenger["Pclass"]==1][mrs_passenger["Embarked"]=="S"]
mrs_2_C = mrs_passenger[mrs_passenger["Pclass"]==2][mrs_passenger["Embarked"]=="C"]
mrs_2_S = mrs_passenger[mrs_passenger["Pclass"]==2][mrs_passenger["Embarked"]=="S"]
mrs_3_C = mrs_passenger[mrs_passenger["Pclass"]==3][mrs_passenger["Embarked"]=="C"]
mrs_3_S = mrs_passenger[mrs_passenger["Pclass"]==3][mrs_passenger["Embarked"]=="S"]
mrs_3_Q = mrs_passenger[mrs_passenger["Pclass"]==3][mrs_passenger["Embarked"]=="Q"]

mrs_1_C["Age"].fillna(mrs_1_C["Age"].mean(), inplace = True)
mrs_1_S["Age"].fillna(mrs_1_S["Age"].mean(), inplace = True)
mrs_2_C["Age"].fillna(mrs_2_C["Age"].mean(), inplace = True)
mrs_2_S["Age"].fillna(mrs_2_S["Age"].mean(), inplace = True)
mrs_3_C["Age"].fillna(mrs_3_C["Age"].mean(), inplace = True)
mrs_3_S["Age"].fillna(mrs_3_S["Age"].mean(), inplace = True)
mrs_3_Q["Age"].fillna(mrs_3_Q["Age"].mean(), inplace = True)

t_mrs_passenger = test_data[test_data["Name"].str.contains("Mrs")]

t_mrs_1_C = t_mrs_passenger[t_mrs_passenger["Pclass"]==1][t_mrs_passenger["Embarked"]=="C"]
t_mrs_1_S = t_mrs_passenger[t_mrs_passenger["Pclass"]==1][t_mrs_passenger["Embarked"]=="S"]
t_mrs_2_C = t_mrs_passenger[t_mrs_passenger["Pclass"]==2][t_mrs_passenger["Embarked"]=="C"]
t_mrs_2_S = t_mrs_passenger[t_mrs_passenger["Pclass"]==2][t_mrs_passenger["Embarked"]=="S"]
t_mrs_3_C = t_mrs_passenger[t_mrs_passenger["Pclass"]==3][t_mrs_passenger["Embarked"]=="C"]
t_mrs_3_S = t_mrs_passenger[t_mrs_passenger["Pclass"]==3][t_mrs_passenger["Embarked"]=="S"]
t_mrs_3_Q = t_mrs_passenger[t_mrs_passenger["Pclass"]==3][t_mrs_passenger["Embarked"]=="Q"]

t_mrs_1_C["Age"].fillna(t_mrs_1_C["Age"].mean(), inplace = True)
t_mrs_1_S["Age"].fillna(t_mrs_1_S["Age"].mean(), inplace = True)
t_mrs_2_C["Age"].fillna(t_mrs_2_C["Age"].mean(), inplace = True)
t_mrs_2_S["Age"].fillna(t_mrs_2_S["Age"].mean(), inplace = True)
t_mrs_3_C["Age"].fillna(t_mrs_3_C["Age"].mean(), inplace = True)
t_mrs_3_S["Age"].fillna(t_mrs_3_S["Age"].mean(), inplace = True)
t_mrs_3_Q["Age"].fillna(t_mrs_3_Q["Age"].mean(), inplace = True)


train_data.update(mrs_1_C)
train_data.update(mrs_2_C)
train_data.update(mrs_1_S)
train_data.update(mrs_2_S)
train_data.update(mrs_3_Q)
train_data.update(mrs_3_C)
train_data.update(mrs_3_S)

test_data.update(t_mrs_1_C)
test_data.update(t_mrs_2_C)
test_data.update(t_mrs_1_S)
test_data.update(t_mrs_2_S)
test_data.update(t_mrs_3_C)
test_data.update(t_mrs_3_S)
test_data.update(t_mrs_3_Q)


## Miss.
ms_passenger = train_data[train_data["Name"].str.contains("Miss")]
ms_passenger["Age"].fillna(ms_passenger["Age"].mean(),inplace = True)
train_data.update(ms_passenger)

t_ms_passenger = test_data[test_data["Name"].str.contains("Miss")]
t_ms_passenger["Age"].fillna(t_ms_passenger["Age"].mean(),inplace = True)
test_data.update(t_ms_passenger)
t_ms__passenger = test_data[test_data["Name"].str.contains("Ms")]
t_ms__passenger["Age"].fillna(t_ms_passenger["Age"].mean(),inplace = True)
test_data.update(t_ms__passenger)
```

```python
from sklearn.ensemble import RandomForestClassifier

y = train_data["Survived"]

features = ["Pclass", "Sex", "SibSp", "Age","Embarked"]

X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])

model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X,y)
predictions = model.predict(X_test)

output = pd.DataFrame({"PassengerID": test_data.PassengerId, "Survived": predictions})
output['PassengerID'] = output['PassengerID'].astype(int)
output['Survived'] = output['Survived'].astype(int)
output.to_csv('my_predictions4.csv', index = False)

print("Successful!")
```
